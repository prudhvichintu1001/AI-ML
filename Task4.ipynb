label encoding
LabelEncoder class can only be used to encode categorical variables, and the price(in lakhs) column is a continuous variable. By encoding the values in the price(in lakhs) column, we can use the LabelEncoder class to transform the data into a format that is more suitable for machine learning algorithms.

```python
!pip install preprocessing
```

***Preprocessing the data takes place***

```python
import sklearn.preprocessing
label_encoder = sklearn.preprocessing.LabelEncoder()
car['registration_year']=label_encoder.fit_transform(car['registration_year'])
car['insurance_validity'] = label_encoder.fit_transform(car['insurance_validity'])
car['fuel_type'] = label_encoder.fit_transform(car['fuel_type'])
car['seats'] = label_encoder.fit_transform(car['seats'])
car['kms_driven'] = label_encoder.fit_transform(car['kms_driven'])
car['ownsership'] = label_encoder.fit_transform(car['ownsership'])
car['transmission'] = label_encoder.fit_transform(car['transmission'])
car['manufacturing_year'] = label_encoder.fit_transform(car['manufacturing_year'])
car['mileage(kmpl)'] = label_encoder.fit_transform(car['mileage(kmpl)'])
car['engine(cc)'] = label_encoder.fit_transform(car['engine(cc)'])
car['max_power(bhp)'] = label_encoder.fit_transform(car['max_power(bhp)'])
car['torque(Nm)'] = label_encoder.fit_transform(car['torque(Nm)'])
car['price(in lakhs)'] = label_encoder.fit_transform(car['price(in lakhs)'])
#car['company'] = label_encoder.fit_transform(car['comapny'])
```

***Correlation matrix***

***The column prints the columns of the new data frame x***

```python
X=car[['registration_year', 'insurance_validity', 'fuel_type',
       'seats', 'kms_driven', 'ownsership', 'transmission',
       'manufacturing_year', 'mileage(kmpl)', 'engine(cc)', 'max_power(bhp)',
       'torque(Nm)']]
Y=car[['price(in lakhs)']]
X.head()
```

***Seaborn's lmplot to create a scatter plot with a linear regression line, showing the relationship between the 'mileage(kmpl)' (mileage) and 'price(in lakhs)' (car price) columns from the 'car' DataFrame***

data splitting

```python
import sklearn.model_selection
x_train,x_test,y_train,y_test =sklearn.model_selection.train_test_split(X,Y,test_size=0.2)
```

***Prints the no of rows and colums in the train and test matrices***

```python
print("x_train:",x_train.shape)
print("x_test:",x_test.shape)
print("y_train:",y_train.shape)
print("y_test:",y_test.shape)
```

***Selected the linear regression model for the model selection***

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

model training

```python
clf.fit(x_train,y_train)
```

model evaluation

```python
clf.score(x_test,y_test)
```

```python
predict=clf.predict(x_test)
predict
```

***Calculating the mean squared error***

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import math
math.sqrt(mean_squared_error(predict,y_test))
```

```python
from sklearn.linear_model import Lasso

# Create a lasso regression model
model = Lasso(alpha=0.2)

# Fit the model to the data
model.fit(x_train,y_train)

# Predict the response for a new data point
y_pred = model.predict(x_test)
y_pred
```

***score method is commonly used to evaluate a model's performance on a test dataset***

```python

model.score(x_test,y_test)
```

***Predict the response for a new data point***

```python
from sklearn.linear_model import Ridge

# Create a lasso regression model
modell = Ridge(alpha=0.2)

# Fit the model to the data
modell.fit(x_train,y_train)

# Predict the response for a new data point
y_pred = modell.predict(x_test)
y_pred
```

***The score method typically takes the input features (x_test) and the corresponding target labels (y_test) as arguments and returns a metric that indicates how well the model performs on the given test data***

```python
modell.score(x_test,y_test)
```
